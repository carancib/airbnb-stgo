```python
## Cargar librerias
import warnings
warnings.filterwarnings(action='once')
import pandas as pd
import numpy as np
airbnb = pd.read_csv('airbnburl.csv', sep=";")
airbnb.head()
```

```python
## Probando leer la url y obtener el html correctamente

test_url = airbnb['url'][1]

import urllib3
hdr = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.11 (KHTML, like Gecko) Chrome/23.0.1271.64 Safari/537.11'}
http = urllib3.PoolManager()

html = http.request('GET', test_url, headers=hdr)
html.data
```

```python
## Cargando soup y encontrando el str que nos interesa scrapear

from bs4 import BeautifulSoup
soup = BeautifulSoup(html.data, 'html.parser')
soup.find_all('span', '_y8ard79')[3].text.replace('\n','')
```

```python
## Crear funcion para extraer cantidad de baños de cada html

def extract_baths(html):    
    soup = BeautifulSoup(html, 'html.parser')
    try:
        return soup.find_all('span', '_y8ard79')[3].text.replace('\n','')
    except:
            return ''
```

```python
## Intentando abrir las primeras 5 urls y guardar a TXT

for url in airbnb['url'][:5]:
    html = http.request('GET', url,  headers=hdr).data
    with open(url+'.txt','x') as f:
        f.write(html)    
        
```
